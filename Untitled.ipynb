```python
## de que nacinoalidad son mayormente los huespedes
nac = df.groupby(['country'])
```

```python
## Los 10 top 10 de reservas por paises, lob que evidencia que probrablemente
## el dataset describe un hotel ubicado ne portugal

nac = nac.size().reset_index(name='counts').nlargest(5,'counts')
```

```python
nac
```

```python
nac.plot(kind='pie', y = 'counts',  autopct='%1.1f%%', 
 startangle=90, shadow=False, labels=nac['country'], legend = False, fontsize=14)
```

```python
## meses donde se reciben mayor numero de reservas


meses = df.groupby(['arrival_date_month'])
```

```python
meses = meses.size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

```python
# Los meses de mayor afluencia de pasajeros corresponden a Agosto Juilo y Mayo
# Correponden con los meses mas calidos del destino
```

```python
## realizo un grafico de barra con los datos obtenidos

```

```python
meses.plot.bar(x='arrival_date_month',y='counts')
```

```python
## Segun evidencia el grafico los meses donde existe un mayor fujo 
## de pasajeros ocurre en los meses mas calidos en el destino
## siendo los meses de menor afluencia de publico los correspondientes 
## al invierno
```
